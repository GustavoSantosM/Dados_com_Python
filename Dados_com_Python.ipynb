```python
column_translations = {
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}

df = df.rename(columns=column_translations)
df.columns
```

```python
df["senioridade"].value_counts()
```

```python
df["contrato"].value_counts()
```

```python
df["remoto"].value_counts()
```

```python
df["tamanho_empresa"].value_counts()
```

```python
senioridade_translations = {
    'SE': 'Sênior',
    'MI': 'Pleno',
    'EN': 'Júnior',
    'EX': 'Executivo'
}

df['senioridade'] = df['senioridade'].replace(senioridade_translations)
df['senioridade'].value_counts()
```

```python
contrato_translations = {
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelancer'
}

df['contrato'] = df['contrato'].replace(contrato_translations)
df['contrato'].value_counts()
```

```python
tamanho_empresa_translations = {
    'M': 'Médio',
    'L': 'Grande',
    'S': 'Pequeno'
}

df['tamanho_empresa'] = df['tamanho_empresa'].replace(tamanho_empresa_translations)
df['tamanho_empresa'].value_counts()
```

```python
remoto_translations = {
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Totalmente Remoto'
}

df['remoto'] = df['remoto'].replace(remoto_translations)
df['remoto'].value_counts()
```

```python
import numpy as np

df_salarios = pd.DataFrame({
  'nome': ["ana", "bruno", "carlos","daniele", "val"],
  'salarios': [4000, np.nan, 5000,np.nan, 10000]
})

df_salarios['salarios_media'] = df_salarios['salarios'].fillna(df_salarios['salarios'].mean().round(2))

df_salarios['salarios_mediana'] = df_salarios['salarios'].fillna(df_salarios['salarios'].median())

df_salarios
```

```python
df_temperatura = pd.DataFrame({
    "dia": ["segunda", "terça", "quarta", "quinta", "sexta"],
    "temperatura": [30, np.nan, np.nan, 28, 27]
})

df_temperatura["preenchido_ffill"] = df_temperatura["temperatura"].ffill()
df_temperatura
```

```python
df_temperatura = pd.DataFrame({
    "dia": ["segunda", "terça", "quarta", "quinta", "sexta"],
    "temperatura": [30, np.nan, np.nan, 28, 27]
})

df_temperatura["preenchido_bfill"] = df_temperatura["temperatura"].bfill()
df_temperatura
```

```python
df_cidades = pd.DataFrame({
    'nome': ["ana", "bruno", "carlos","daniele", "val"],
    'cidades': ["são paulo", np.nan, "curitiba", np.nan, "belén"]
})

df_cidades['cidade_preenchida'] = df_cidades["cidades"].fillna("não informado")
df_cidades
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df_limpo, x='senioridade', y='usd')
plt.title("salario medio por senioridade")
plt.xlabel("senioridade")
plt.ylabel("salario medio anual (USD)")
plt.show()
```

```python
df_limpo.groupby('senioridade')['usd'].mean().sort_values(ascending=False)
```

```python
ordem = df_limpo.groupby('senioridade')['usd'].mean().sort_values(ascending=True).index
```

```python
ordem
```

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df_limpo, x='senioridade', y='usd', order=ordem)
plt.title("salario medio por senioridade")
plt.xlabel("senioridade")
plt.ylabel("salario medio anual (USD)")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df_limpo['usd'], bins = 50, kde=True)
plt.title("distribuição dos salarios anuais")
plt.xlabel("salarios em usd")
plt.ylabel("frequência")
plt.show()
```
